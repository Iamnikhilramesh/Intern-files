```python
#define lists to collect the data
sku_id = []
property_name = []
#define headers
headers = {
    "Accept" : "application/json",
    "Content-Type" : "application/json",
    'Authorization': 'WM3 NUpSNkp4VFB4WGpYbzBYenp1TmdMUQ=='
}
stop_words = ["visibility", "information_provider", "last_synchronized", "subproduct_ean",
                  "statistics_code","product_name","target_market",
                  "net_content","descriptor_code","last_time_of_change","trade_item_size",
                  "nutrient_comparison_type","minimum_trade_item_lifespan_from_time_of_production","brand_name",
                  "gpc_category_name","comparison_per_litre","packaging_marked_label_accreditation_code",
                  "validoo_country_of_origin","comparison_per_kilogram","comparison_drained_weight","comparison_per_piece",
                  "subproduct_ean","minimum_lifespan","marketing_message","import_classification","nutritional_claim",
                 "comparison_","packaging_marked_diet_allergen_code","Etikettnamn","bast_fore_datum"]
#API call
for j in sku_list:
    #API call
    r = requests.get('https://klippkungen-new.wm3.se/api/v1/shop/admin/products?includes[]=product_properties&sku={sku_id}'.format(sku_id = j),headers=headers)    
    if r.status_code == 200:
        raw_data = r.json()
        if raw_data["products"] != []:
            if raw_data["products"][0]['product_properties'] is not None:
                temp_list = []
                for i in raw_data["products"][0]['product_properties']:
                    if i['property_name'] not in stop_words:
                        temp_list.append(i['property_name']+'-'+i['value'])
                property_name.append(temp_list)
                sku_id.append(j)
```

# Convert to dataframe

```python
#convert the exracted data to list and then to dataframe
l = [sku_id,property_name]
col_names = ['sku_id','property_name']
df = pd.DataFrame(l).transpose()
df.columns = col_names
df
```
